```r
ipr_scan <- read.csv('../vitvi.vcostv3.clean.pep.txt', sep='\t')

## need to filter for genes tested in DGE analysis
ipr_sub <- ipr_scan %>% 
  select(genename, e.value, ipr_accession) %>%
  filter(e.value < 1e-10) %>% 
  filter(ipr_accession != '') %>%
  select(genename, ipr_accession) %>%
  unique()

#also grab pfam
pfam_sub <- ipr_scan %>% 
  select(genename, analysis, signature_accession, signature_description, e.value, ipr_accession) %>%
  filter(e.value < 1e-10) %>% 
  filter(analysis == 'Pfam') %>%
  select(genename, signature_accession, signature_description, ipr_accession) %>%
  unique()
```

```r
full_enrichment_test <- function(set_of_genes, ipr_sub, print_res=FALSE){
    ipr_testset <- ipr_sub[ipr_sub$genename %in% set_of_genes,]
    terms <- unique(ipr_testset$ipr_accession)
    
    print(str_interp("analyzing ${length(terms)} IPR accessions..."))
    
    enriched_terms <- c() 
    for (term in terms){
        
        genes_to_term_in_set <- nrow(ipr_testset[ipr_testset$ipr_accession == term,])
        genes_to_term_in_genome <- nrow(ipr_sub[ipr_sub$ipr_accession == term,])
        genes_not_term_in_genome <- 24769 - genes_to_term_in_genome
        size_of_set <- length(set_of_genes)
         
        test <- phyper(genes_to_term_in_set, genes_to_term_in_genome,
                       genes_not_term_in_genome, size_of_set, lower.tail=FALSE)
        
        if (test < 0.05/length(terms) & 
            genes_to_term_in_genome >= 10 & 
            genes_to_term_in_set >= 3){
            if (print_res == TRUE) {
                print(c(term, test, genes_to_term_in_set, genes_to_term_in_genome))
            }
            enriched_terms <- c(enriched_terms, term, genes_to_term_in_set)
        }
    }
    
    return(enriched_terms)

}

full_pfam_enrichment_test <- function(set_of_genes, pfam_sub, print_res=FALSE){
    pfam_testset <- pfam_sub[pfam_sub$genename %in% set_of_genes,]
    terms <- unique(pfam_testset$signature_accession)
    
    print(str_interp("analyzing ${length(terms)} Pfam domains..."))
    
    enriched_terms <- c()
    for (term in terms){
        
        genes_to_term_in_set <- nrow(pfam_testset[pfam_testset$signature_accession == term,])
        genes_to_term_in_genome <- nrow(pfam_sub[pfam_sub$signature_accession == term,])
        genes_not_term_in_genome <- 24769 - genes_to_term_in_genome
        size_of_set <- length(set_of_genes)
        
        test <- phyper(genes_to_term_in_set, genes_to_term_in_genome,
                       genes_not_term_in_genome, size_of_set, lower.tail=FALSE)
        
        if (test < 0.05/length(terms) & 
            genes_to_term_in_genome >= 10 & 
            genes_to_term_in_set >= 3){
            if (print_res == TRUE){
                print(c(term, test, genes_to_term_in_set, genes_to_term_in_genome))
            }
            enriched_terms <- c(enriched_terms, term, genes_to_term_in_set)
        }
    }
    
    return(enriched_terms)

}
```

```r
## make whole analysis into a function

functional_enrichment <- function(gPC, ipr_scan, ipr_sub, pfam_sub){
    
    # get high loading genes
    high_genes <- read.csv(str_interp('genePC_loaders/${gPC}_high_loaders.csv'))[, 'x']
    high_genes <- paste(high_genes, '.t01', sep='')
    
    # get low loading genes
    low_genes <- read.csv(str_interp('genePC_loaders/${gPC}_low_loaders.csv'))[, 'x']
    low_genes <- paste(low_genes, '.t01', sep='')
    
    all_genes <- read.csv('all_rootstock_comparisons/Leaf_2018_Harvest_Ungrafted-1103P.csv')[, 'X']
    all_genes <- paste(all_genes, '.t01', sep='')
    
    #filter for genes in DEG analysis
    ipr_sub_mtv <- ipr_sub %>%
      filter(genename %in% all_genes)

    pfam_sub_mtv <- pfam_sub %>%
      filter(genename %in% all_genes)

    #run ipr and pfam enrichments - high loaders
    eset_up <- full_enrichment_test(high_genes, ipr_sub_mtv)
    pset_up <- full_pfam_enrichment_test(high_genes, pfam_sub_mtv)
   
    #run ipr and pfam enrichments - low loaders
    eset_down <- full_enrichment_test(low_genes, ipr_sub_mtv)
    pset_down <- full_pfam_enrichment_test(low_genes, pfam_sub_mtv)   
    
    
    #make output tables
    ipr_up <- ipr_scan %>%
      select(ipr_accession, ipr_description) %>%
      unique() %>%
      filter(ipr_accession %in% eset_up)#[1])
    
    pfam_up <- ipr_scan %>%
      filter(analysis == 'Pfam') %>%
      select(signature_accession, signature_description) %>%
      unique() %>%
      filter(signature_accession %in% pset_up)#[1])
    
    ipr_down <- ipr_scan %>%
      select(ipr_accession, ipr_description) %>%
      unique() %>%
      filter(ipr_accession %in% eset_down)#[1])
    
    pfam_down <- ipr_scan %>%
      filter(analysis == 'Pfam') %>% 
      select(signature_accession, signature_description) %>%
      unique() %>%
      filter(signature_accession %in% pset_down)#[1])
    
    output <- list()
    output[['full_eset_up']][['enriched_terms']] <- eset_up#[1]
    #output[['full_eset_up']][['genes_to_term_in_set']] <- eset_up[2]
    output[['full_pset_up']][['enriched_terms']] <- pset_up#[1]
    #output[['full_pset_up']][['genes_to_term_in_set']] <- pset_up[1]
    output[['full_eset_down']][['enriched_terms']] <- eset_down#[1]
    #output[['full_eset_down']][['genes_to_term_in_set']] <- eset_down[2]
    output[['full_pset_down']][['enriched_terms']] <- pset_down#[1]
    #output[['full_pset_down']][['genes_to_term_in_set']] <- pset_down[2]
    
    output[['pretty_eset_up']][['enriched_terms']] <- ipr_up#[1]
    #output[['pretty_eset_up']][['genes_to_term_in_set']] <- ipr_up[2]
    output[['pretty_pset_up']][['enriched_terms']] <- pfam_up#[1]
    #output[['pretty_pset_up']][['genes_to_term_in_set']] <- pfam_up[2]
    output[['pretty_eset_down']][['enriched_terms']] <- ipr_down#[1]
    #output[['pretty_eset_down']][['genes_to_term_in_set']] <- ipr_down[2]
    output[['pretty_pset_down']][['enriched_terms']] <- pfam_down#[1]
    #output[['pretty_pset_down']][['genes_to_term_in_set']] <- pfam_down[2]
    
    return(output)
}
```

```r
## heavy PC loaders

make_pretty_output <- function(gPC){
    x <- functional_enrichment(gPC, ipr_scan=ipr_scan, ipr_sub=ipr_sub, pfam_sub=pfam_sub)
    high_e <- x[['pretty_eset_up']][['enriched_terms']]
    colnames(high_e) <- c('term', 'desc')
    high_p <- x[['pretty_pset_up']][['enriched_terms']]
    colnames(high_p) <- c('term', 'desc')
    low_e <- x[['pretty_eset_down']][['enriched_terms']]
    colnames(low_e) <- c('term', 'desc')
    low_p <- x[['pretty_pset_down']][['enriched_terms']]
    colnames(low_p) <- c('term', 'desc')
    x <- bind_rows(high_e, high_p, low_e, low_p)
    x$direction <- c(rep('high', nrow(high_e)+nrow(high_p)), rep('low', nrow(low_e)+nrow(low_p)))
    x$PC <- rep(gPC, nrow(x))
    return(x)
}

pcs <- c('PC226', 'PC241', 'PC186', 'PC267', 
         'PC105', 'PC183', 'PC202', 'PC97', 
         'PC151', 'PC201', 'PC233', 'PC277')

for (pc in pcs){
    if (pc == pcs[1]){
        df <- make_pretty_output(pc)
    } else {
        x <- make_pretty_output(pc)
        df <- rbind(df, x)
    }
}
```

```r
plotvenn <- function(rs1103p, rs3309c, rsso4, fname, fnameformat){
    # change the name of the pdf saved based on what kind of venn is being generated
    fileprefix= "/data/projects/julia.pratt/CS1_genomeSelection/figs/go/venn/venn_pset_"
    f <- paste(fileprefix, fname, '.png', sep="")
    
    # the function takes in three lists
    # automatically counts how many are shared/private within those lists
    # most of this is just styling
    venn.diagram(
        x = list(rs1103p, rs3309c, rsso4),
        
        cex = 0.5,
        fontfamily = "sans",
        
        category.names = c("1103P", "3309C", "SO4"),
        cat.cex = 0.5,
        cat.default.pos = "outer",
        cat.dist = c(0.075, 0.075, 0.075),
        cat.fontfamily = "sans",
        
        main = fnameformat,
        main.cex = 0.6,
        main.fontfamily = "sans",
        
        col=c("#1b9e77", "#7570b3", "#e6ab02"),
        fill = c(alpha("#1b9e77",0.3), alpha("#7570b3",0.3), alpha("#e6ab02",0.3)),
        
        filename = f,
        output=TRUE, 
        resolution = 300,
        imagetype="png",
        margin=0.075,
        rotation = 1,
        height=700,
        width=700
    )
}
```
